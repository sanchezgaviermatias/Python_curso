## Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from pandas_datareader import data
plt.style.use('bmh')
```

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

## Histogramas

```python
df1.head()
```

```python
df1['A'].plot.hist() #Variables Continuas
```

```python
df1['A'].plot.hist(bins=50) 
```

## Tipos de Gráficos

Hay varios métodos construidos en pandas para hacer gráficos estadísticos: 
    
* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

***

## Gráfico de Área 

```python
df2.head()
```

```python
df2.plot.area(alpha=0.4, cmap='coolwarm')
```

## Bar Plot

```python
df2.head()
```

```python
df2.plot.bar() #Variables Categóricas
```

## Line Plots 

```python
df1.head()
```

```python
df1[["A","B"]].plot.line(figsize=(30,10))
```

```python
df1["A"].plot.line(ls='-.', c='red', lw='0.5',figsize=(25,5), title="Precio de Acciones").set(xlabel="Fecha", ylabel="Precios");
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

Se pude usar c para indicar el color en base a otra variable.
Cmap indica el colormap a usar. 
Para ver todos los colormap visita: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

## Box plot

```python
df2.plot.box() # x categórica
```

```python
df2["a"].plot.box()
```

## Hexagonal Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

## Kernel Density Estimation plot (KDE)

```python
df2.plot.density()
plt.savefig('books_read2.png') #Guardar figura
```
